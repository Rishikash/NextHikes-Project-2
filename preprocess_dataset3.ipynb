# Boxplot for atemp

```python
sns.boxplot(x=merged_dataset3['atemp'])
plt.title('Boxplot of Atemp')
plt.show()
```

```python
Q1=0.166700
Q3=0.257600
Lowerlimit = Q1-1.5*(Q3-Q1)
Upperlimit = Q3+1.5*(Q3-Q1)
merged_dataset3['atemp'] = np.where(merged_dataset3['atemp']>Upperlimit,merged_dataset3['atemp'].mean(),merged_dataset3['atemp'])
merged_dataset3['atemp'] = np.where(merged_dataset3['atemp']<Lowerlimit,merged_dataset3['atemp'].mean(),merged_dataset3['atemp'])
```

```python
sns.boxplot(x=merged_dataset3['atemp'])
plt.title('Boxplot of Atemp')
plt.show()
```

# Boxplot for atemp, hum, and windspeed

# Boxplot of cnt grouped by mnth

```python
sns.boxplot(x='season', y='cnt', data=merged_dataset3)
plt.title('Boxplot of Total Users by season')
plt.show()
```

```python
q1, q3 = np.percentile(merged_dataset3['cnt'], [25, 75])
```

```python
Llimit = q1-1.5*(q3-q1)
Uimit = q3+1.5*(q3-q1)
merged_dataset3['cnt'] = np.where(merged_dataset3['cnt']>Uimit,merged_dataset3['cnt'].median(),merged_dataset3['cnt'])
```

```python
sns.boxplot(x='season', y='cnt', data=merged_dataset3)
plt.title('Boxplot of Total Users by season')
plt.show()
```
